```python
import numpy as np
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as mp

warnings.filterwarnings("ignore")
mp.rcParams['font.family'] = ['sans-serif']
mp.rcParams['font.sans-serif'] = ['SimHei']
mp.rcParams['axes.unicode_minus']=False

funds = pd.read_csv('funds_details.csv',sep=',',na_values='')
df_funds = pd.DataFrame(funds)
df_funds.iloc[57:65,:]
```

```python
for item in df_funds.columns:
    df_funds[item]=df_funds[item].apply(lambda x: np.NaN if str(x)=='' else x)
```

```python
df_funds.info()
```

```python
df_funds = df_funds.dropna(how='any',axis=0)
df_funds.isnull().sum().sort_values(ascending=False)
```

```python
strange_place_list = []
for item in df_funds.columns:
    strange_place_list.append(df_funds[df_funds[item]=='--']['fnum'].count())
print(strange_place_list)
```

```python
strange_index_list = []
unique_strange_set = set()
for item in ['cnv_change_6m','unit_nv_change','npv_thisy']:
    strange_index_list.append(df_funds[df_funds[item]=='--'].index.tolist())
for item in strange_index_list:
    for num in item:
        unique_strange_set.add(num)
unique_strange = list(unique_strange_set)
print(len(unique_strange))
for item in unique_strange:
    df_funds = df_funds.drop(item)
```

```python
strange_place_list = []
for item in df_funds.columns:
    strange_place_list.append(df_funds[df_funds[item]=='--']['fnum'].count())
print(strange_place_list)
```

```python
df_funds.reset_index(drop=True, inplace=True)
df_funds.head(10)
```

```python
df_funds.to_csv('funds_sorted.csv',index=False)
```

```python
funds_sorted = pd.DataFrame(pd.read_csv('funds_sorted.csv'))
funds_sorted.head(7)
```

```python
for item in ['cumulative_nv','cnv_change_6m','unit_nv_change','pnv_change_1m','npv_thisy','fee']:
    df_funds[item] = df_funds[item].str.replace('%','')
    df_funds[item] = df_funds[item].astype('float')
    df_funds[item] = df_funds[item].apply(lambda x : x/100)
df_funds.info()
```

```python
# for item in ['class','risk','management','manager']:
#     print(df_funds[item].value_counts(ascending=False))
fig,axes = mp.subplots(nrows=2,ncols=2,figsize=(13,9))
mp.subplots_adjust(left=0.15,bottom=0.1,top=1.2,right=0.95,hspace=0.2,wspace=0.25)
c = 0
r = 0
for item in ['class','manager','risk','management']:
    if c<= 1:
        ax = sns.barplot(x=df_funds[item].value_counts().index.tolist()[:10],y=df_funds[item].value_counts().tolist()[:10],ax=axes[r,c],hue_order=df_funds[item].value_counts().index.tolist()[:10])
        ax.set_title(item,fontsize=14)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45,fontsize=12)
        c += 1
    else:
        r += 1
        c = 0
        ax = sns.barplot(x=df_funds[item].value_counts().index.tolist()[:10],y=df_funds[item].value_counts().tolist()[:10],ax=axes[r,c],hue_order=df_funds[item].value_counts().index.tolist()[:10])
        ax.set_title(item,fontsize=14)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45,fontsize=12)
        c += 1   
```

```python
df_funds['scale'] = df_funds['scale'].str.split('亿').str[0]
df_funds['scale'] = df_funds['scale'].astype('float')
```

```python
df_funds['mmin_amount'] = df_funds['mmin_amount'].str.split('元').str[0]
df_funds['mmin_amount'] = df_funds['mmin_amount'].astype('float')
```

```python
print(df_funds[df_funds['rank_1y'].str.contains('--')]['fnum'].count())
strange_index_list = []
strange_index_list.append(df_funds[df_funds['rank_1y'].str.contains('--')].index.tolist())
for item in strange_index_list:
    df_funds = df_funds.drop(item)
df_funds.reset_index(drop=True, inplace=True)
print(df_funds[df_funds['rank_1y'].str.contains('--')]['fnum'].count())
```

```python
dived = df_funds['rank_1y'].str.split('|').str[0].astype('float')
div = df_funds['rank_1y'].str.split('|').str[1].astype('float')
df_funds['rank_1y'] = dived/div
```

```python
# print(pd.to_datetime(df_funds['founded_date']).dt.to_period('M'))
# pd.to_datetime(df_funds['founded_date']).dt.to_period('Y')
year = df_funds['founded_date'].str.split('-').str[0]
month = df_funds['founded_date'].str.split('-').str[1]
df_funds['founded_date'] = (year+month).astype('float')
```

```python
import sklearn.preprocessing as sp
encoder_dict = {}
for item in ['class','manager','risk','management']:
    lbe = sp.LabelEncoder()
    df_funds[item] = lbe.fit_transform(df_funds[item])
    encoder_dict[item] = lbe
```

```python
print(encoder_dict['class'].inverse_transform([6])[0])
print(encoder_dict['manager'].inverse_transform([326])[0])
print(encoder_dict['risk'].inverse_transform([2])[0])
print(encoder_dict['management'].inverse_transform([65])[0])
```

```python
df_funds.index = df_funds['fnum']
df_funds.head()
```

```python
scmat = []
for i,fund_a in df_funds.iterrows():
#     print(fund_a[2:])
    scrow = []
    for j,fund_b in df_funds.iterrows():
        score = np.corrcoef(fund_a[2:].astype('float64'),fund_b[2:].astype('float64'))[0, 1]
        scrow.append(score)
    scmat.append(scrow)
```

```python
df_score = pd.DataFrame(scmat)
df_score.index = df_funds['fnum']
df_score.columns = df_funds['fnum']
df_score
```

```python
df_score.to_csv('score.csv')
```
